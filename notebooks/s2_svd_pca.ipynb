```python
import numpy as np

X = np.array([
    [2.0,  4.1, 1.0],
    [3.0,  5.9, 1.8],
    [4.0,  8.2, 2.1],
    [5.0,  9.8, 3.2],
    [6.0, 12.1, 3.9],
    [7.0, 13.9, 4.2]
])

# 1. Centering
mean = X.mean(axis=0)
X_centered = X - mean

# 2. Reduced SVD
U, S, Vt = np.linalg.svd(X_centered,full_matrices=False)

# 3. Exact SVD reconstruction
Sigma = np.diag(S)
X_reconstructed = U @ Sigma @ Vt

print(np.allclose(X_centered, X_reconstructed))  # True

# 4. Explained variance
energy = S ** 2
explained_variance_ratio = energy / energy.sum()

print("Singular values:", S)
print("Explained variance ratio:", explained_variance_ratio)

# 5. Rank-2 truncated SVD

k = 2
U_k = U[:, :k]
S_k = S[:k]
Vt_k = Vt[:k, :]

X_rank_k = U_k @ np.diag(S_k) @ Vt_k

# 6. Approximation error
measured_error = (np.linalg.norm(X_centered - X_rank_k, ord="fro") ** 2)

theoretical_error = np.sum(S[k:] ** 2)

print("Measured error:", measured_error)
print("Theoretical error:", theoretical_error)

# 7. PCA projection
V_k = Vt_k.T
Z = X_centered @ V_k

print("PCA representation shape:", Z.shape)  # (6, 2)

# 8. PCA reconstruction
X_from_pca = Z @ V_k.T

print(np.allclose(X_from_pca, X_rank_k))  # True
```
